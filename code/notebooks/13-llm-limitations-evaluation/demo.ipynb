# Chapter 13 — LLM Limitations & Evaluation: calibration (companion demo)

Deterministic, no network. Reproduces the Expected Calibration Error worked example
(`ex:ch13-ece-credit`, ECE = 0.115) and the reliability diagram
(`fig:ch13-reliability`). The full evaluation lab is in `exercises.ipynb`.

```python
import numpy as np
import matplotlib.pyplot as plt

# Fixed bin statistics from ex:ch13-ece-credit (5 equal-width bins).
conf  = np.array([0.08, 0.31, 0.50, 0.71, 0.88])
acc   = np.array([0.11, 0.28, 0.47, 0.55, 0.62])
count = np.array([120, 180, 200, 300, 200])

ece = float(np.sum(count / count.sum() * np.abs(conf - acc)))
print(f'ECE = {ece:.3f}')  # 0.115
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.plot([0, 1], [0, 1], '--', color='gray', label='perfect calibration')
ax.bar(conf, acc, width=0.16, color='#4c72b0', edgecolor='black', alpha=0.85,
       label='model accuracy')
for c, a in zip(conf, acc):
    ax.plot([c, c], [a, c], color='#d62728')
ax.set_xlim(0, 1); ax.set_ylim(0, 1); ax.set_aspect('equal')
ax.set_xlabel('confidence'); ax.set_ylabel('accuracy')
ax.set_title(f'Reliability diagram (ECE={ece:.3f})'); ax.legend()
plt.show()
```
