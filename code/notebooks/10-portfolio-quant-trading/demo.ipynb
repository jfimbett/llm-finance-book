# Chapter 10 — Portfolio & Quant Trading: efficient frontier (companion demo)

Deterministic, no network. Reproduces the mean-variance efficient frontier
(`fig:ch10-frontier`) for a fixed 4-asset example and compares the tangency portfolio
with the 1/N rule. The full backtest lab is in `exercises.ipynb`.

```python
import numpy as np
import matplotlib.pyplot as plt

mu = np.array([0.008, 0.010, 0.006, 0.012])
Sigma = np.array([[0.0040, 0.0010, 0.0008, 0.0015],
                  [0.0010, 0.0055, 0.0012, 0.0020],
                  [0.0008, 0.0012, 0.0030, 0.0009],
                  [0.0015, 0.0020, 0.0009, 0.0070]])
inv = np.linalg.inv(Sigma); ones = np.ones(4)
A = ones @ inv @ ones; B = ones @ inv @ mu; C = mu @ inv @ mu; D = A*C - B*B

targets = np.linspace(mu.min()-0.001, mu.max()+0.002, 200)
std = np.sqrt(np.maximum((A*targets**2 - 2*B*targets + C)/D, 0))

w_tan = inv @ mu / (ones @ inv @ mu)
w_eq = ones/4
for name, w in [('tangency', w_tan), ('1/N', w_eq)]:
    m, s = w @ mu, np.sqrt(w @ Sigma @ w)
    print(f'{name:9s} mu={m:.4f} sd={s:.4f} Sharpe={m/s:.3f}')
```

```python
plt.plot(std*100, targets*100, label='efficient frontier')
for name, w, mk in [('tangency', w_tan, 'o'), ('1/N', w_eq, 's')]:
    m, s = w @ mu, np.sqrt(w @ Sigma @ w)
    plt.scatter([s*100], [m*100], marker=mk, label=name, zorder=5)
plt.xlabel('volatility (%/month)'); plt.ylabel('expected excess return (%/month)')
plt.legend(); plt.title('Mean-variance frontier'); plt.show()
```
