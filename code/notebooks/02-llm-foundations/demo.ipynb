# Chapter 2 — LLM Foundations: sinusoidal positional encoding (companion demo)

This notebook reproduces the positional-encoding figure (`fig:ch02-illustration`)
deterministically (no data or network) and checks that cosine similarity between
position encodings decays with distance. The same logic lives in
`gen_positional_encoding.py`; the full lab is in `exercises.ipynb`.

```python
import numpy as np
import matplotlib.pyplot as plt

def positional_encoding(seq_len, d_model):
    pos = np.arange(seq_len)[:, None]
    i = np.arange(d_model)[None, :]
    rates = 1.0 / np.power(10000.0, (2 * (i // 2)) / d_model)
    ang = pos * rates
    pe = np.zeros((seq_len, d_model))
    pe[:, 0::2] = np.sin(ang[:, 0::2])
    pe[:, 1::2] = np.cos(ang[:, 1::2])
    return pe

pe = positional_encoding(50, 64)
plt.imshow(pe, cmap='RdBu', aspect='auto', origin='lower')
plt.xlabel('Encoding dimension'); plt.ylabel('Token position'); plt.colorbar()
plt.show()
```

```python
def cos(a, b):
    return float(a @ b / (np.linalg.norm(a) * np.linalg.norm(b)))

print(f"cos(PE_5, PE_10) = {cos(pe[5], pe[10]):.3f}")
print(f"cos(PE_5, PE_45) = {cos(pe[5], pe[45]):.3f}")
# Nearby positions are more similar than distant ones.
```
