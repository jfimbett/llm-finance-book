# Chapter 12 — Explainability: SHAP token attributions (companion demo)

This notebook reproduces the SHAP attribution figure (`fig:ch12-shap`) for the
credit-decision classifier of Example *SHAP Attribution for a Credit Decision*.
It is fully deterministic and uses no live model or data: the token-level SHAP
values are the fixed values stated in the chapter text. The same logic lives in
`gen_shap_attribution.py`.

```python
import matplotlib.pyplot as plt

# Fixed SHAP values from the chapter example (baseline 0.50 -> model output 0.14).
tokens = [
    ("struggled", -0.089),
    ("declining", -0.073),
    ("down 30%", -0.061),
    ("competitive", -0.047),
    ("restaurant", -0.038),
    ("pandemic", -0.029),
    ("operates", +0.012),
]
print(f"sum of shown token contributions = {sum(v for _, v in tokens):+.3f}")
```

```python
items = sorted(tokens, key=lambda t: t[1])
labels = [t[0] for t in items]
values = [t[1] for t in items]
colors = ["#d62728" if v < 0 else "#2ca02c" for v in values]

fig, ax = plt.subplots(figsize=(7.2, 4.2))
ax.barh(range(len(values)), values, color=colors, edgecolor="black", linewidth=0.4)
ax.set_yticks(range(len(values)))
ax.set_yticklabels(labels)
ax.axvline(0.0, color="black", linewidth=0.8)
ax.set_xlabel("SHAP value (contribution to P(creditworthy))")
ax.set_title("Token attributions for a credit-decision classifier")
plt.tight_layout()
plt.show()
```

The negative (red) tokens account for most of the `0.50 - 0.14 = 0.36` gap between
the baseline and the model's output, while `operates` makes a small positive
contribution. These attributions are what an ECOA adverse-action notice would be
built from. See `exercises.ipynb` for the full lab.
